```python
%matplotlib inline
import numpy as np
from astropy.table import Table
import matplotlib.pyplot as plt
```

```python
# parameters for plot
fs = 16 # fontsize
lw = 2  # linewidth
ms = 4  # marksize
pa = 0.8 # alpha of points
ea = 0.1 # alpha of errorbar


fig, ax = plt.subplots(1,1, figsize=(10, 8))
ax.tick_params(axis='both', which='major', labelsize=fs)
ax.tick_params(axis='both', which='minor', labelsize=fs-1)

# default error, use for fitting
x_err_default = 0.2*np.log(10) # 20% error
y_err_default = 0.2*np.log(10) # 20% error

if True: # plot data of this work
    manga_data = Table.read('manga.txt', format='ascii')
    manga_L = manga_data['log(L[OIII])']
    manga_L_err = manga_data['log(L[OIII])_err']
    manga_r = manga_data['log(R16)']
    manga_r_err = manga_data['log(R16)_err']
    is_limit = (manga_data['is_limit'] == 1 )
    
    color = 'C0'
    marker = 'o'
    ax.plot(manga_L, manga_r, marker=marker, alpha=pa, label='This Work', 
            color=color, lw=lw, ms=ms, linestyle='None')
    plt.errorbar(manga_L, manga_r, xerr=manga_L_err, uplims=is_limit,
                 yerr=manga_r_err, ls='None', label=None, alpha=ea, color=color)
    

if True: # plot fraqulli 2003
    fra03_data = Table.read('Fraquelli2003.txt', format='ascii')
    fra03_L = fra03_data['log(L[OIII])']
    fra03_L_err = fra03_data['log(L[OIII])_err']
    fra03_r = fra03_data['log(R16)']
    fra03_r_err = fra03_data['log(R16)_err']
    is_limit = (fra03_data['is_limit'] == 0 )
    color = 'C1'
    marker = 'x'
    ax.plot(fra03_L, fra03_r, marker=marker, alpha=pa, label='Fraquelli 2003', color=color, 
            lw=lw, ms=ms, linestyle='None')
    ax.errorbar(fra03_L, fra03_r, marker=marker, alpha=ea, label=None, color=color,
                linestyle='None')


if True: # plot Bennert 2006a
    ben06a_data = Table.read('Bennert2006a.txt', format='ascii')
    ben06a_L = ben06a_data['log(L[OIII])']
    ben06a_L_err = ben06a_data['log(L[OIII])_err']
    ben06a_r = ben06a_data['log(R16)']
    ben06a_r_err = ben06a_data['log(R16)_err']
    is_limit = (ben06a_data['is_limit'] == 0 )
    color = 'C2'
    marker = '>'
    ax.plot(ben06a_L, ben06a_r, marker=marker, alpha=pa, label='Bennert 2006a', 
            color=color, lw=lw, ms=ms, linestyle='None')
    plt.errorbar(ben06a_L, ben06a_r, ls='None', label=None, alpha=ea, color=color,)

if True: # plot Bennert 2006b
    ben06b_data = Table.read('Bennert2006b.txt', format='ascii')
    ben06b_L = ben06b_data['log(L[OIII])']
    ben06b_L_err = ben06b_data['log(L[OIII])_err']
    ben06b_r = ben06b_data['log(R16)']
    ben06b_r_err = ben06b_data['log(R16)_err']
    is_limit = (ben06b_data['is_limit'] == 0 )
    
    color = 'C3'
    marker = '<'
    ax.plot(ben06b_L, ben06b_r, marker=marker, alpha=pa, label='Bennert 2006b', 
            color=color, lw=lw, ms=ms, linestyle='None')
    plt.errorbar(ben06b_L, ben06b_r, ls='None', label=None, alpha=ea, color=color,)

    
if True: # plot greene 2011
    gre11_data = Table.read('Greene2011.txt', format='ascii')
    gre11_L = gre11_data['log(L[OIII])']
    gre11_L_err = gre11_data['log(L[OIII])_err']
    gre11_r = gre11_data['log(R16)']
    gre11_r_err = gre11_data['log(R16)_err']
    is_limit = (gre11_data['is_limit'] == 0 )
    color = 'C4'
    marker = 'P'
    ax.plot(gre11_L, gre11_r, marker=marker, alpha=pa, label='Greene 2011', 
            color=color, lw=lw, ms=ms, linestyle='None')
    plt.errorbar(gre11_L, gre11_r, xerr=gre11_L_err, yerr=gre11_r_err, 
                 ls='None', label=None, alpha=ea, color=color)

    
if True: # plot liu2013
    liu13_data = Table.read('Liu2013.txt', format='ascii')
    liu13_L = liu13_data['log(L[OIII])']
    liu13_L_err = liu13_data['log(L[OIII])_err']
    liu13_r = liu13_data['log(R16)']
    liu13_r_err = liu13_data['log(R16)_err']
    is_limit = (liu13_data['is_limit'] == 0 )
    color = 'C5'
    marker = 's'
    ax.plot(liu13_L, liu13_r, marker=marker, alpha=pa, label='Liu 2013', 
            color=color, lw=lw, ms=ms, linestyle='None')
    plt.errorbar(liu13_L, liu13_r, yerr=liu13_r_err, xerr=liu13_L_err,
                 ls='None', label=None, alpha=ea, color=color)
    
if True: # plot liu2014
    liu14_data = Table.read('Liu2014.txt', format='ascii')
    liu14_L = liu14_data['log(L[OIII])']
    liu14_L_err = liu14_data['log(L[OIII])_err']
    liu14_r = liu14_data['log(R16)']
    liu14_r_err = liu14_data['log(R16)_err']
    is_limit = (liu14_data['is_limit'] == 0 )
    
    color = 'C6'
    marker = 'D'
    ax.plot(liu14_L, liu14_r, marker=marker, alpha=pa, label='Liu 2014', 
            color=color, lw=lw, ms=ms, linestyle='None')
    plt.errorbar(liu14_L, liu14_r, yerr=liu14_r_err, xerr=liu14_L_err,
                 ls='None', label=None, alpha=ea, color=color)


if True: # plot Law 2018
    law18_data = Table.read('Law2018.txt', format='ascii')
    law18_L = law18_data['log(L[OIII])']
    law18_L_err = law18_data['log(L[OIII])_err']
    law18_r = law18_data['log(R16)']
    law18_r_err = law18_data['log(R16)_err']
    is_limit = (law18_data['is_limit'] == 0 )
    color = 'C7'
    marker = '*'
    ax.plot(law18_L, law18_r, marker=marker, alpha=pa, label='Law 2018', color=color, 
            lw=lw, ms=ms, linestyle='None')
    ax.errorbar(law18_L, law18_r, marker=marker, alpha=ea, label=None, color=color,
                linestyle='None')


    
ax.set_xlabel(r"${\rm Log(L[O\,III])}$", fontsize=fs)
ax.set_ylabel(r"Log(R$_{16}$)/pc", fontsize=fs)
ax.set_xlim(39, 44.5)
ax.set_ylim(2.3, 4.8)
ax.legend(fontsize=fs, loc='upper left')
#fig.savefig('./all_fitting.pdf', bbox_inches='tight')
```
